```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')#used to ignore warnings
```

```python
df=pd.read_csv("Melbourne_housing_FULL.csv")
```

```python
cols_to_use=['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
```

```python
df=df[cols_to_use]
```

```python
df
```

```python
df.isna().sum()
```

```python
cols_to_fill_zero=['Propertycount','Distance','Bedroom2','Bathroom','Car']
```

```python
df.loc[:,cols_to_fill_zero]=df.loc[:,cols_to_fill_zero].fillna(0)
```

```python
df.isna().sum()
```

```python
df.loc[:,'Landsize']=df.loc[:,'Landsize'].fillna(df.Landsize.mean())
df.loc[:,'BuildingArea']=df.loc[:,'BuildingArea'].fillna(df.Landsize.mean())
# df.loc[:,'Price']=df.loc[:,'Price'].fillna(df.Price.mean())
```

```python
 df.isna().sum()
```

```python
df.dropna()
```

```python
df.isna().sum()
```

```python
dataset=pd.get_dummies(df,drop_first=True).astype(int)
dataset.head()
```

```python
X=dataset.drop('Price',axis=1)
y=dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
train_X,test_x,train_Y,test_y=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regLinear=LinearRegression().fit(train_X,train_Y)
```

```python
regLinear
```

```python
regLinear.score(test_x,test_y)
```

```python
regLinear.score(train_X,train_Y)
```

```python
from sklearn.linear_model import Lasso
lasso_reg=Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_reg.fit(train_X,train_Y)
```

```python
lasso_reg.score(test_x,test_y)
```

```python
from sklearn.linear_model import Ridge
ridge_reg=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_reg.fit(train_X,train_Y)
```
